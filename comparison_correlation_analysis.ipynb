```python
pdb_num = pdb.select_dtypes(include=[pd.np.number])
scaler = MinMaxScaler()
pdb_normalize = pd.DataFrame(scaler.fit_transform(pdb_num), columns=pdb_num.columns)
spearman_corr_pdb = pdb_normalize.corr(method='spearman')
threshold = 0.7
high_spearman_corr_pdb = np.where(np.abs(spearman_corr_pdb) > threshold)
spearman_pairs_pdb = [(spearman_corr_pdb.index[x], spearman_corr_pdb.columns[y]) for x, y in zip(*high_spearman_corr_pdb) if x != y]

af2_num = af2.select_dtypes(include=[pd.np.number])
af2_normalize = pd.DataFrame(scaler.fit_transform(af2_num), columns=af2_num.columns)
spearman_corr_af2 = af2_normalize.corr(method='spearman')
high_spearman_corr_af2 = np.where(np.abs(spearman_corr_af2) > threshold)
spearman_pairs_af2 = [(spearman_corr_af2.index[x], spearman_corr_af2.columns[y]) for x, y in zip(*high_spearman_corr_af2) if x != y]

pisces_num = pisces.select_dtypes(include=[pd.np.number])
pisces_normalize = pd.DataFrame(scaler.fit_transform(pisces_num), columns=pisces_num.columns)
spearman_corr_pisces = pisces_normalize.corr(method='spearman')
high_spearman_corr_pisces = np.where(np.abs(spearman_corr_pisces) > threshold)
spearman_pairs_pisces = [(spearman_corr_pisces.index[x], spearman_corr_pisces.columns[y]) for x, y in zip(*high_spearman_corr_pisces) if x != y]

pdb_af2_num = pdb_af2.select_dtypes(include=[pd.np.number])
pdb_af2_normalize = pd.DataFrame(scaler.fit_transform(pdb_af2_num), columns=pdb_af2_num.columns)
spearman_corr_pdb_af2 = pdb_af2_normalize.corr(method='spearman')
high_spearman_corr_pdb_af2 = np.where(np.abs(spearman_corr_pdb_af2) > threshold)
spearman_pairs_pdb_af2 = [(spearman_corr_pdb_af2.index[x], spearman_corr_pdb_af2.columns[y]) for x, y in zip(*high_spearman_corr_pdb_af2) if x != y]

pisces_af2_num = pisces_af2.select_dtypes(include=[pd.np.number])
pisces_af2_normalize = pd.DataFrame(scaler.fit_transform(pisces_af2_num), columns=pisces_af2_num.columns)
spearman_corr_pisces_af2 = pisces_af2_normalize.corr(method='spearman')
high_spearman_corr_pisces_af2 = np.where(np.abs(spearman_corr_pisces_af2) > threshold)
spearman_pairs_pisces_af2 = [(spearman_corr_pisces_af2.index[x], spearman_corr_pisces_af2.columns[y]) for x, y in zip(*high_spearman_corr_pdb_af2) if x != y]

```

```python
# Convert the lists of pairs to sets
spearman_pairs_pdb_set = set(spearman_pairs_pdb)
spearman_pairs_af2_set = set(spearman_pairs_af2)
spearman_pairs_pisces_set = set(spearman_pairs_pisces)
spearman_pairs_pdb_af2_set = set(spearman_pairs_pdb_af2)
spearman_pairs_pisces_af2_set = set(spearman_pairs_pisces_af2)

# Find overlapping pairs
overlapping_pairs = spearman_pairs_pdb_set.intersection(spearman_pairs_af2_set, spearman_pairs_pisces_set, spearman_pairs_pdb_af2_set, spearman_pairs_pisces_af2_set)
print("Overlapping pairs:")
for pair in overlapping_pairs:
    print(pair)

# Find unique pairs for each dataset
unique_pairs_pdb = spearman_pairs_pdb_set.difference(overlapping_pairs)
unique_pairs_af2 = spearman_pairs_af2_set.difference(overlapping_pairs)
unique_pairs_pisces = spearman_pairs_pisces_set.difference(overlapping_pairs)
unique_pairs_pdb_af2 = spearman_pairs_pdb_af2_set.difference(overlapping_pairs)
unique_pairs_pisces_af2 = spearman_pairs_pisces_af2_set.difference(overlapping_pairs)

# Print unique pairs for each dataset
print("\nUnique pairs in pdb:")
for pair in unique_pairs_pdb:
    print(pair)

print("\nUnique pairs in af2:")
for pair in unique_pairs_af2:
    print(pair)

print("\nUnique pairs in pisces:")
for pair in unique_pairs_pisces:
    print(pair)

print("\nUnique pairs in pdb_af2:")
for pair in unique_pairs_pdb_af2:
    print(pair)

print("\nUnique pairs in pisces_af2:")
for pair in unique_pairs_pisces_af2:
    print(pair)
```
